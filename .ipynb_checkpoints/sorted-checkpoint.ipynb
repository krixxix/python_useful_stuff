# Sorted
- this function will sort an iterable object in ascending or descending order 
- depends on our demands


```python
numbers = [4,5,2,3,-1,0,9]
```

#### Example 1

```python
sorted_nums = sorted(numbers)
print(sorted_nums)
```

#### Example 2

```python
sorted_nums2 = sorted(numbers, reverse=True)
print(sorted_nums2)
```

#### Example 3

```python
people = [
    {"name": "Alice", "age": 30},
    {"name": "Bob", "age": 25},
    {"name": "Charlie", "age":35},
    {"name": "David", "age": 20},
]

sorted_people = sorted(people, key=lambda person: person["age"])
print(sorted_people)
```

# Statsmodels
- scikit-learn: machine learning focus (prediction first)
- statsmodels: statistical modeling focus (inference first)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
from statsmodels.formula.api import ols, logit

np.random.seed(42)
```

```python
# -----------------------------
# Create Example Dataset
# -----------------------------
n = 200

df = pd.DataFrame({
    "age": np.random.randint(20, 60, n),
    "experience": np.random.randint(1, 20, n),
    "education_years": np.random.randint(10, 20, n),
})

# Create salary with signal + noise
df["salary"] = (
    2000
    + 500 * df["experience"]
    + 300 * df["education_years"]
    + 50 * df["age"]
    + np.random.normal(0, 5000, n)
)

df
```

## Ordinary Least Squares (OLS)

```python
model = ols("salary ~ age + experience + education_years", data=df)
results = model.fit()

print(results.summary())
```

### Access Important Components

```python
results.params        # Coefficients
results.pvalues       # p-values
results.rsquared      # R²
results.resid         # Residuals
results.fittedvalues  # Predictions on training data
```

### Predictions

```python
df["predicted_salary"] = results.predict(df)
```

### Confidence Intervals

```python
results.conf_int()
```

### Interaction Terms

```python
model = ols("salary ~ experience * education_years", data=df)
results = model.fit()

print(results.summary())
```

### Categorical Variables
- ``C()`` tells satsmodels to treat variable as categorical (creates dummy variables automatically)

```python
df["department"] = np.random.choice(["IT", "HR", "Sales"], n)

model = ols("salary ~ experience + C(department)", data=df)
results = model.fit()

print(results.summary())
```

## Logistic Regression

```python
# Create binary outcome
df["high_income"] = (df["salary"] > df["salary"].median()).astype(int)
```

```python
# fit logistic model
log_model = logit("high_income ~ experience + education_years", data=df)
log_results = log_model.fit()

print(log_results.summary())
```

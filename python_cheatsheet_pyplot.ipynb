# Python Cheatsheet: pyplot

```python
import matplotlib.pyplot as plt  # Main plotting library
import numpy as np               # Numerical operations
```

## Basic Plot

```python
x = [1, 2, 3, 4]              # X values
y = [10, 20, 25, 30]          # Y values

plt.plot(x, y)                # Create line plot
plt.show()                    # Render the plot window
```

## Line Plot Customization
markers:
- ``o``: circle
- ``s``: square
- ``^``: triangle
- ``x``: x-mark
- ``*``: star

line styles:
- ``-``: solid
- ``--``: dashed
- ``:``: dotted
- ``-.``: dash-dot

```python
plt.plot(
    x, y,
    color="blue",             # Line color
    linestyle="--",           # Dashed line
    linewidth=2,              # Line thickness
    marker="o",               # Point markers
    markersize=6,             # Marker size
    label="Line 1"            # Legend label
)

plt.title("My Plot")          # Plot title
plt.xlabel("X Axis")          # X-axis label
plt.ylabel("Y Axis")          # Y-axis label
plt.legend()                  # Show legend
plt.grid(True)                # Show grid
plt.show()
```

## Scatter Plot

```python
plt.scatter(x, y)          # Basic scatter plot
plt.show()

plt.scatter(
    x, y,
    c="red",               # Point color
    s=100,                 # Point size
    alpha=0.7              # Transparency (0–1)
)
plt.show()
```

## Bar Charts

### Vertical Bar

```python
categories = ["A", "B", "C"]   # Categories
values = [5, 7, 3]             # Heights

plt.bar(categories, values)    # Vertical bar chart
plt.show()
```

### Horizontal Bar

```python
plt.barh(categories, values)   # Horizontal bar chart
plt.show()
```

## Histogram

```python
data = np.random.randn(1000)   # Generate random data

plt.hist(data, bins=30)        # Histogram with 30 bins
plt.show()
```

## Boxplot

### Basic Boxplot


```python
# Generate sample data
data = np.random.randn(1000)   # Normally distributed data

# Basic boxplot
plt.boxplot(data)
plt.title("Basic Boxplot")
plt.show()
```

### Customized Boxplot

```python
plt.boxplot(
    data,
    vert=True,                 # Vertical orientation (False = horizontal)
    patch_artist=True,         # Enable box fill color
    showmeans=True,            # Show mean marker
    showfliers=True,           # Show outliers
    notch=True                 # Notched box (confidence interval around median)
)

plt.title("Customized Boxplot")
plt.ylabel("Values")
plt.grid(True)
plt.show()
```

## Subplots (Object-Oriented API)

```python
fig, ax = plt.subplots(2, 2, figsize=(8, 6))  # 2x2 grid of plots

ax[0, 0].plot(x, y)           # Top-left plot
ax[0, 1].scatter(x, y)        # Top-right plot
ax[1, 0].bar(categories, values)
ax[1, 1].hist(data)

plt.tight_layout()            # Prevent overlapping
plt.show()
```

## Figure Control

```python
plt.figure(figsize=(10, 5))   # Set figure size (width, height in inches)
plt.plot(x, y)
plt.show()

plt.savefig(
    "plot.png",
    dpi=300,                  # High resolution
    bbox_inches="tight"       # Remove extra whitespace
)
```

## Axis Control

```python
plt.plot(x, y)

plt.xlim(0, 10)               # Set X-axis limits
plt.ylim(0, 40)               # Set Y-axis limits

plt.xticks([1, 2, 3, 4])      # Custom X ticks
plt.yticks([10, 20, 30])      # Custom Y ticks

plt.show()
```

## Styles
common styles:
- ``ggplot``
- ``seaborn-v0_8``
- ``fivethirtyeight``
- ``dark_background``
- ``bmh``
- ...

```python
plt.style.use("ggplot")       # Apply built-in style

plt.plot(x, y)
plt.show()
```

## Annotations

```python
plt.plot(x, y)

plt.annotate(
    "Important Point",        # Text label
    xy=(2, 20),               # Arrow target
    xytext=(3, 25),           # Text location
    arrowprops=dict(facecolor="black")  # Arrow style
)

plt.show()
```

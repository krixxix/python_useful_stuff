# Python Cheatsheet: seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# -----------------------------
# Create Example Dataset
# -----------------------------
np.random.seed(42)

df = pd.DataFrame({
    "age": np.random.randint(20, 60, 100),
    "salary": np.random.normal(50000, 15000, 100),
    "department": np.random.choice(["IT", "HR", "Sales"], 100),
    "experience": np.random.randint(1, 20, 100),
    "performance_score": np.random.normal(70, 10, 100)
})

# Set nice default styling
sns.set_style("whitegrid")
```

```python
df.head()
```

```python
df.shape
```

## Scatterplot

```python
sns.scatterplot(
    data=df,
    x="age",
    y="salary",
    hue="department"  # Automatically color groups
)

plt.title("Age vs Salary by Department")
plt.show()
```

## Lineplot

```python
sns.lineplot(
    data=df,
    x="experience",
    y="salary"
)

plt.title("Experience vs Salary")
plt.show()
```

- Groups the data by x
- Calculates the mean of y for each x
- Draws a line through those means
- Calculates a 95% confidence interval around those means
- Shades that region

## Countplot

```python
sns.countplot(
    data=df,
    x="department"
)

plt.title("Department Counts")
plt.show()
```

## Barplot

```python
sns.barplot(
    data=df,
    x="department",
    y="salary"
    #,estimator=np.median
    #,errorbar = None
    #,errorbar = ('ci', 99)
)

plt.title("Average Salary per Department")
plt.show()
```

## Boxplot

```python
sns.boxplot(
    data=df,
    x="department",
    y="salary"
)

plt.title("Salary Distribution by Department")
plt.show()
```

## Violinplot

```python
sns.violinplot(
    data=df,
    x="department",
    y="salary"
)

plt.title("Salary Distribution Shape by Department")
plt.show()
```

## Histogram

```python
sns.histplot(
    data=df,
    x="salary",
    bins=20,
    kde=True  # Adds smooth density curve
)

plt.title("Salary Distribution")
plt.show()
```

## KDE Plot

```python
sns.kdeplot(
    data=df,
    x="salary",
    fill=True
)

plt.title("Salary Density Curve")
plt.show()
```

## Heatmap

```python
correlation = df.corr(numeric_only=True)

sns.heatmap(
    correlation,
    annot=True,     # Show numbers
    cmap="coolwarm"
)

plt.title("Correlation Matrix")
plt.show()
```

## Pairplot

```python
sns.pairplot(
    df,
    hue="department"
)

plt.show()
```
